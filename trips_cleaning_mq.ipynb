```python
import pandas as pd
import json 
```

```python
stations = {}
with open("stations.json", "r") as file:
    stations = json.load(file)
```

```python
df_02_2024 = pd.read_csv("./bikes_data/trips_data/202402-bluebikes-tripdata.csv")
df_02_2024["started_at"] = pd.to_datetime(df_02_2024["started_at"],
                            format="%Y-%m-%d %H:%M:%S")
df_02_2024["ended_at"] = pd.to_datetime(df_02_2024["ended_at"],
                            format="%Y-%m-%d %H:%M:%S")
```

```python
def add_month_of_rides(df, rides, subscriber, start_time, end_time, start_station, end_station, 
                       subscribe_convention, bike_type=None):
    df = df.dropna(subset=[start_station, end_station])
    df[subscriber] = df[subscriber].apply(lambda sub: sub == subscribe_convention).astype(int).astype(float)  # Binary: subbed or no? 1=yes, 0=no

    for i in range(len(df)):
        row = df.iloc[i]
        temp = {
            "subscribed": row[subscriber],
            "start_time":row[start_time],
            "end_time":row[end_time],
            "start_station": stations[row[start_station]],
            "end_station": stations[row[end_station]]
        }

        if bike_type is not None:
            temp["bike_type"] = row[bike_type]
        
        rides.append(temp)

        if i%1000 == 0:
            print("finished: " + str(i))
    return rides
```

```python
trips_array = []
trips_array = add_month_of_rides(df_02_2024, trips_array, "member_casual", "started_at", "ended_at", "start_station_name", "end_station_name",
                   "member", "rideable_type")
print(trips_array)
```
